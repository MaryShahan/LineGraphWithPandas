# Sketch the graph with some modification

```python
import pandas as pd

y = [2,4,6,8]
x = [1,3,5,7]

graph = pd.DataFrame(y , x)
graph.plot(kind = 'line')
```

```python
y = {'f(x)' : [2,4,6,8]}
x = [1,3,5,7]

graph = pd.DataFrame(y , x)
graph.plot(kind = 'line')
```

```python
y = {'f(x)' : [2,4,6,8],
     'g(x)' : [4,-2,1,5]}
x = [1,3,5,7]

graph = pd.DataFrame(y , x)
graph.plot(kind = 'line', grid=True, title='<<>>My Graph<<>>', ylabel='Y-Mary-Y', xlabel='X-Mary-X')
```
